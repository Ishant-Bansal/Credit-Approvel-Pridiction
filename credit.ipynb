```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
data = pd.read_csv('credit.csv')
```

```python
df = pd.DataFrame(data)
df.head(3)
```

```python
df.dtypes 
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.dtypes
```

```python
df['duration'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
le = LabelEncoder()
df['job'] = le.fit_transform(df['job'])
```

```python
df['housing'] = le.fit_transform(df['housing'])
df['education'] = le.fit_transform(df['education'])
df['marital'] = le.fit_transform(df['marital'])
```

```python
ohe = OneHotEncoder()
df['approval']= ohe.fit_transform(df[['approval']]).toarray()
```

```python
df
```

```python
x = df.drop(['approval'], axis=1)
y = df['approval']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
input_shape = x_train.shape[1]
input_shape
```

```python
model = Sequential()
model.add(Dense(units = 128,activation='relu' , input_dim = input_shape))

model.add(Dense(units = 64 , activation='relu'))
model.add(Dense(units = 32 , activation='relu'))
model.add(Dense(units = 16 , activation='relu'))
model.add(Dense(units = 8 , activation='relu'))

model.add(Dense(units= 1,activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['mae', 'mse','accuracy'])
```

```python
model.fit(x_train,y_train,epochs=10,batch_size=32)
```

```python
pred = np.ravel(model.predict(x_test))
pred = pd.DataFrame(pred)
```

```python
pred =(pred>=0.5).astype(int)
```
